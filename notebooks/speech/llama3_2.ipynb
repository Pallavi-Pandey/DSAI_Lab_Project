```python
import subprocess

def ask_ollama(model, prompt):
    """
    Sends a prompt to an Ollama model and returns the response.
    """
    result = subprocess.run(
        ["ollama", "run", model],
        input=prompt,
        text=True,
        capture_output=True
    )
    return result.stdout.strip()


def generate_medical_response(user_type, user_input):
    """
    Generates different responses depending on whether the user is
    a doctor or a patient.
    """

    # ---- PROMPTS FOR DIFFERENT USERS ----
    doctor_prompt = f"""
You are a highly trained clinical medical assistant for DOCTORS.
Use clinical terminology, differential diagnosis, and medical reasoning.

Doctor Input:
{user_input}

Provide:
- Differential diagnoses
- Clinical interpretation
- Suggested tests or next steps
- Professional and concise medical language
    """

    patient_prompt = f"""
You are a friendly medical assistant for PATIENTS.
Explain medical concepts in simple, easy-to-understand language.

Patient Input:
{user_input}

Provide:
- Simple explanation
- Possible causes
- What they can do now
- Friendly, reassuring tone
    """

    # Select the correct prompt
    if user_type.lower() == "doctor":
        final_prompt = doctor_prompt
    elif user_type.lower() == "patient":
        final_prompt = patient_prompt
    else:
        return "Error: user_type must be 'doctor' or 'patient'."

    # ---- CALL OLLAMA ----
    output = ask_ollama("llama3.2", final_prompt)
    return output


# -------------- MAIN PROGRAM -------------------

user_type = input("Are you a doctor or a patient? ").strip()
user_input = input("Enter your message: ").strip()

response = generate_medical_response(user_type, user_input)

print("\n--- AI Response ---")
print(response)
```
